```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw =2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.ylim(-2,2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta1 = 30
theta2 = 0.5

x = np.arange(0,100)
y = theta1+theta2*x

plt.plot(x,y)
plt.xlim(0,50)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta1 = 25
theta2 = 20

x = np.arange(-100,150)
y = (x - theta1)**2 + theta2

plt.annotate('local min', xy=(25, 20), xytext=(30, 2500),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3,3,0.1)
temp = np.log10(abs(x))
y = -1*temp

plt.plot(x,y)
plt.ylim(-2,6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3,3,0.1)
temp = np.log10(abs(1-x))
y = -1*temp

plt.plot(x,y)
plt.ylim(-2,6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2,2,0.1)
y = 1/(1+np.exp(-x))

plt.plot(x,y)
plt.ylim(0,1)
plt.show()
```
